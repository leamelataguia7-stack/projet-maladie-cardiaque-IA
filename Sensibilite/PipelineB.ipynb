```python
#Construction du Pipeline B (BRUT)
import pandas as pd

df = pd.read_csv(r"C:\Users\YOGA Slim\Documents\GitHub\projet-maladie-cardiaque-IA\BASES_Données\BASE_EDA\FRAMINGANG.csv", sep=";"  )
print(df.shape)

df.columns = [
    "sexe", "age", "niveau_etude", "fumeur", "cigarettes_par_jour",
    "traitement_hypertenseur", "antecedent_avc", "hypertension", "diabete",
    "cholesterol_total", "tension_sys", "tension_dia", "imc",
    "frequence_cardiaque", "glucose", "maladie_cardiaque_10ans"
]

df.head()
```

```python
#Renommer les colonnes
df.columns = [
    "sexe", "age", "niveau_etude", "fumeur", "cigarettes_par_jour",
    "traitement_hypertenseur", "antecedent_avc", "hypertension", "diabete",
    "cholesterol_total", "tension_sys", "tension_dia", "imc",
    "frequence_cardiaque", "glucose", "maladie_cardiaque_10ans"
]

df.head()
df.info()
```

```python
#Création de age_cat (variable descriptive)
df["age_cat"] = pd.cut(
    df["age"],
    bins=[0, 40, 60, 80, 120],
    labels=["<40", "40-60", "60-80", ">80"],
    right=False
)

df["age_cat"].value_counts()
```

```python
#Définition des variables quantitatives / qualitatives
target = "maladie_cardiaque_10ans"

categorical_vars = [
    "sexe", "niveau_etude", "fumeur", "traitement_hypertenseur",
    "antecedent_avc", "hypertension", "diabete"
]

quantitative_vars = [
    "age", "cigarettes_par_jour", "cholesterol_total",
    "tension_sys", "tension_dia", "imc",
    "frequence_cardiaque", "glucose"
]
```

```python
#Gestion des valeurs manquantes(imputation par la médiane pour les quantitatives et par la modalité la plus fréquente pour les qualitatives     )
from sklearn.impute import SimpleImputer

imputer_quant = SimpleImputer(strategy="median")
imputer_cat   = SimpleImputer(strategy="most_frequent")

df[quantitative_vars] = imputer_quant.fit_transform(df[quantitative_vars])
df[categorical_vars]  = imputer_cat.fit_transform(df[categorical_vars])

df.isnull().sum()
```

```python
#exclure valeur manquante sur la variable cible 
df = df[df[target].notna()]
df.isnull().sum()
```

```python
df.columns
```

```python
#Définir les variables catégorielles et quantitatives

categorical_vars_B = [
    "sexe", "niveau_etude", "fumeur",
    "traitement_hypertenseur", "antecedent_avc",
    "hypertension", "diabete"
]

quantitative_vars_B = [
    "age", "cigarettes_par_jour", "cholesterol_total",
    "tension_sys", "tension_dia", "imc",
    "frequence_cardiaque", "glucose"
]

```

```python
#Split TRAIN / TEST
from sklearn.model_selection import train_test_split

X_B = df.drop(columns=["maladie_cardiaque_10ans"])
y_B = df["maladie_cardiaque_10ans"]

X_train_B, X_test_B, y_train_B, y_test_B = train_test_split(
    X_B, y_B,
    test_size=0.2,
    random_state=42,
    stratify=y_B
)
X_train_B.shape, X_test_B.shape
```

```python
#Encodage One-Hot
X_train_B_enc = pd.get_dummies(
    X_train_B,
    columns=categorical_vars_B,
    drop_first=True
)

X_test_B_enc = pd.get_dummies(
    X_test_B,
    columns=categorical_vars_B,
    drop_first=True
)

# Alignement
X_train_B_enc, X_test_B_enc = X_train_B_enc.align(
    X_test_B_enc,
    join="left",
    axis=1
)

# Remplir les colonnes absentes dans le test
X_test_B_enc = X_test_B_enc.fillna(0)

X_train_B_enc.head(), X_test_B_enc.head()   
```

```python
#Gestion du déséquilibre de la variable cible avec SMOTE vs NON-SMOTE. (Synthetic Minority Over-sampling Technique)
```

```python
#Version NON–SMOTE (Train original)
from imblearn.over_sampling import SMOTE

# Initialisation du SMOTE
smote = SMOTE(random_state=42)

# Application du SMOTE sur TRAIN seulement
X_train_B_smote, y_train_B_smote = smote.fit_resample(
    X_train_B_enc,
    y_train_B
)

# Vérification du nouvel équilibre
print("Avant SMOTE :", y_train_B.value_counts(normalize=True) * 100)
print("\nAprès SMOTE :", y_train_B_smote.value_counts(normalize=True) * 100)

# Vérification des dimensions
X_train_B_enc.shape, X_train_B_smote.shape
```

```python
X_train_B_enc[quantitative_vars_B].describe().T
```

```python
#Standardisation
```

```python
#Standardisation du TRAIN NON SMOTE

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit sur TRAIN NON-SMOTE
scaler.fit(X_train_B_enc[quantitative_vars_B])

# Création des versions scalées
X_train_B_scaled = X_train_B_enc.copy()
X_test_B_scaled = X_test_B_enc.copy()

# Application du scaling
X_train_B_scaled[quantitative_vars_B] = scaler.transform(X_train_B_enc[quantitative_vars_B])
X_test_B_scaled[quantitative_vars_B]  = scaler.transform(X_test_B_enc[quantitative_vars_B])
```

```python
#Standardisation du TRAIN SMOTE
scaler_smote = StandardScaler()

# Fit sur TRAIN SMOTE
scaler_smote.fit(X_train_B_smote[quantitative_vars_B])

# Création des versions scalées
X_train_B_smote_scaled = X_train_B_smote.copy()
X_test_B_smote_scaled  = X_test_B_enc.copy()

# Application du scaling
X_train_B_smote_scaled[quantitative_vars_B] = scaler_smote.transform(X_train_B_smote[quantitative_vars_B])
X_test_B_smote_scaled[quantitative_vars_B]  = scaler_smote.transform(X_test_B_enc[quantitative_vars_B])
```

```python
# Verification des dimensions finales
print("A1 - RAW :", X_train_B_enc.shape)
print("A2 - SCALED :", X_train_B_scaled.shape)

print("B1 - SMOTE RAW :", X_train_B_smote.shape)
print("B2 - SMOTE SCALED :", X_train_B_smote_scaled.shape)

print("TEST RAW :", X_test_B_enc.shape)
print("TEST SCALED :", X_test_B_scaled.shape)
print("TEST SMOTE SCALED :", X_test_B_smote_scaled.shape)
```
